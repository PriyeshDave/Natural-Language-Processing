## Hyper-parameter tuning - advance_xgb

```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import GridSearchCV, RepeatedStratifiedKFold, train_test_split
```

```python
dataset = pd.read_csv('../Datasets/Advance-train.csv')
```

```python
model = joblib.load('../Model/xgb_advance.pkl')
```

```python
dataset.head()
```

```python
dataset.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
dataset.head()
```

```python
X_train_advance, X_test_advance, y_train_advance, y_test_advance = train_test_split(dataset.drop(['is_duplicate'], axis=1), dataset['is_duplicate'], test_size=.2, random_state=2)
```

```python
cv = RepeatedStratifiedKFold(n_repeats=5, n_splits=5, random_state=42)
```

```python
param_grid = {'objective': ['binary:logistic'],
 'colsample_bylevel': [.5,1],
 'colsample_bytree': [.5,1],
 'gamma': [0, .1],
 'learning_rate': [0.05,0.1, 0.5],
 'max_depth': [6, 8, 12],
 'min_child_weight': [.5,1],
 'n_estimators': [100, 200, 300],
 'n_jobs': [12],
 'random_state': [42],
 'reg_alpha': [0,.05,.1],
 'reg_lambda': [.05,1],
 'subsample': [.5,1],
}
```

```python
param_grid = {
 'colsample_bylevel': [1],
 'colsample_bytree': [1],
 'learning_rate': [0.1],
 'max_depth': [6],
 'min_child_weight': [1],
 'n_estimators': [200],
 'n_jobs': [12],
 'random_state': [42],
}
```

```python
grid_search = GridSearchCV(model, param_grid= param_grid, cv=cv, n_jobs=8)
```

```python
grid_search.fit(X_train_advance, y_train_advance)
```
